```python
import numpy as np
import scipy.stats as st


def find_point_estimate(data):
    print("Точечные оценки")
    mean_val = np.mean(data)
    print(f"Выборочное среднее = {mean_val:.4f}")
    var_val = np.var(data, ddof=1)
    std_val = np.std(data, ddof=1)
    print(f"Исправленная дисперсия (s²) = {var_val:.4f}")
    print(f"Исправленное СКО (s) = {std_val:.4f}")
```

```python
data = np.array([5, 3, 7, 6, 4, 5, 8, 5, 4, 3])
find_point_estimate(data)
```

```python
x = np.array([1, 3, 6, 26])
freq = np.array([8, 40, 10, 2])
n = np.sum(freq)

data = []
for x, f in zip(x, freq):
    data.extend([x] * f)

find_point_estimate(data)
```

```python
x = np.array([2560, 2600, 2620, 2650, 2700])
freq = np.array([2, 3, 10, 4, 1])
n = np.sum(freq)

data = []
for x, f in zip(x, freq):
    data.extend([x] * f)

find_point_estimate(data)
```

```python
def confidence_interval_norm(data, sigma, gamma):
    n = len(data)
    mean_val = np.mean(data)
    z_krit = st.norm.ppf((1 + gamma) / 2)
    margin = z_krit * sigma / np.sqrt(n)
    return mean_val - margin, mean_val + margin


def confidence_interval_student(data, sigma, gamma):
    n = len(data)
    mean_val = np.mean(data)
    t_krit = st.t.ppf((1 + gamma) / 2, df=n - 1)
    margin = t_krit * sigma / np.sqrt(n)
    return mean_val - margin, mean_val + margin
```

```python
data = np.array([10.1, 9.9, 10.3, 9.7])
n = len(data)
sigma = 0.2
gamma = 0.95

mean_val = np.mean(data)

lower, upper = confidence_interval_norm(data, sigma, gamma)

print(f"95% доверительный интервал: ({lower:.4f}; {upper:.4f})")

lower, upper = st.norm.interval(
    confidence=gamma, loc=mean_val, scale=sigma / np.sqrt(n)
)

print(f"95% доверительный интервал: ({lower:.4f}; {upper:.4f})")
```

```python
data = np.array([1.2, 0.2, 1.6, 1.9, -0.4, 0.5, 1.4, 1.7, 2.5, 1.3])
n = len(data)
gamma = 0.95

mean_val = np.mean(data)
std_val = np.std(data, ddof=1)

lower, upper = confidence_interval_student(data, std_val, gamma)

print(f"95% доверительный интервал: ({lower:.4f}; {upper:.4f})")

lower, upper = st.t.interval(
    confidence=gamma, df=n - 1, loc=mean_val, scale=st.sem(data)
)

print(f"95% доверительный интервал: ({lower:.4f}; {upper:.4f})")
```
